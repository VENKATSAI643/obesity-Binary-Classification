```python
df['NObeyesdad'] = df['NObeyesdad'].map({
    'Obesity_Type_I': 1,
    'Obesity_Type_III': 1,
    'Obesity_Type_II': 1,
    'Overweight_Level_II': 1,
    'Overweight_Level_I': 1,
    'Insufficient_Weight': -1,
    'Normal_Weight': 0
}).fillna(-1)
```

```python
df = pd.get_dummies(df, columns=['CALC', 'Gender', 'FAVC', 'SCC', 'SMOKE', 'family_history_with_overweight', 'CAEC', 'MTRANS'], drop_first=True)
```

```python
df=df.drop(columns=['Height','Age'])
```

```python
df = df.astype(int)
df.head()
```

```python
X= df.drop(columns=['NObeyesdad'])
y=df['NObeyesdad']
from sklearn.model_selection import train_test_split
X_train,X_test ,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model= Sequential()

model.add(Dense(3,activation='sigmoid',input_dim=21))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train,epochs=50,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
y_log=model.predict(X_test_scaled)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
